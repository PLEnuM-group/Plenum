# Define the three different singal models

```python
phi100s = 10**np.linspace(-23,-17, 20)
gammas = np.arange(1.5,4.1,0.25)

# single power law model
# free parameters: phi0, gamma
single_pl_model = SinglePowerLaw(phi0_bins=phi100s, gamma_bins=gammas,
                             n_fitbins_phi0=150, n_fitbins_gamma=80)

# two component power law model
# free parameters: phi0, gamma0, gamma1
E_th = 5e4
two_comp_model = TwoComponentPowerLaw(phi0_bins=phi100s, gamma0_bins=gammas, gamma1_bins=gammas,
                                      E_threshold=E_th, n_fitbins_phi0=60, n_fitbins_gamma0=40,
                                      n_fitbins_gamma1=40)

# power law model with exponential cutoff
# free parameters: phi0, gamma, cutoff energy
cutoff_energy_bins = 10**np.linspace(5,9,15)
cutoff_pl_model = CutOffPowerLaw(phi0_bins=phi100s, gamma_bins=gammas, 
                                 cutoff_energy_bins= cutoff_energy_bins,
                                 n_fitbins_phi0=60, n_fitbins_gamma=30,
                                 n_fitbins_cutoff=20)
```

# Get the background distribution for an analysis that compares a two component power-law to a single power-law
the idea is that we run trials for both hypothesis vs a pure atm background hypothesis and then substract the 
ts values to get the desired test statistic value of this test
Note: the data within the actual background hypothesis should contain a single power-law astro component

```python
#phi0_exp, gamma_exp = 1.44e-18, 2.28
n_trials = int(50)

N = 10

pbar = ProgressBar(N,parent=None).start()
# reset random number generator to guarantee same trials
ana_single_pl.reset_random_number_generator(seed=0)
ana_two_comp_pl.reset_random_number_generator(seed=0)

astro_bckg=False

for k in range(N):

    if k ==0:
        bckg_trials = ana_single_pl.do_trials(n_trials=n_trials, 
                               bckg_kwargs =dict(atm_keys=['numu_conv'], astro_bckg=astro_bckg), integrate=True)
        bckg_trials_two_comp = ana_two_comp_pl.do_trials(n_trials=n_trials, 
                               bckg_kwargs =dict(atm_keys=['numu_conv'], astro_bckg=astro_bckg), integrate=True)
    else:
        bckg_i = ana_single_pl.do_trials(n_trials=n_trials, 
                               bckg_kwargs =dict(atm_keys=['numu_conv'], astro_bckg=astro_bckg), integrate=True)
        bckg_two_comp_i = ana_two_comp_pl.do_trials(n_trials=n_trials, 
                               bckg_kwargs =dict(atm_keys=['numu_conv'], astro_bckg=astro_bckg), integrate=True)
        for exp_key in bckg_trials.keys():
            bckg_trials[exp_key] = np.append(bckg_trials[exp_key], bckg_i[exp_key])
            bckg_trials_two_comp[exp_key] = np.append(bckg_trials_two_comp[exp_key], 
                                                      bckg_two_comp_i[exp_key])
    pbar.increment()

pbar.finish()
```

```python
fig, ax = plt.subplots()

ts_bins = np.linspace(0,5,25)
for exp_key, trials in bckg_trials.items():
    trials2 = bckg_trials_two_comp[exp_key]
    #print(trials2['ts']-trials['ts'])
    ax.hist(trials2['ts']-trials['ts'], bins=ts_bins, histtype='step', label=exp_key, 
            density=True)
     
    xvals = get_mids(ts_bins)
    fit_func = delta_chi2(eta=0.5, df=1.)
    fit_vals = fit_func.pdf(xvals)
    fit_vals[:] /= (np.sum(fit_vals) /(fit_func.eta))
    ax.plot(xvals, fit_vals / np.diff(xvals)[0] , color='black', lw=2., 
                label=r'$0.50\cdot \chi^2$(dof=1)')
    
ax.set_xlabel('test statistic value')
ax.set_ylabel('pdf')

ax.legend()
ax.set_yscale('log')

```

# Do some trials with events injected from a two component power-law


```python
n_trials = int(50)

N = 10

ppbar = ProgressBar(len(g_inj0s),parent=None).start()
# reset random number generator to guarantee same trials
ana_single_pl.reset_random_number_generator(seed=0)
ana_two_comp_pl.reset_random_number_generator(seed=0)

astro_bckg=False
phi0_exp, gamma_exp = 1.44e-18, 2.28


RES = dict()
for i, g_0i in enumerate(g_inj0s):
    phi0_inji = PHI_injs[i]

    params= [phi0_inji, g_0i, gamma_exp]

    pbar = ProgressBar(N,parent=ppbar).start()
    for k in range(N):

        if k ==0:
            data, inj_trials_two_comp = ana_two_comp_pl.do_trials(n_trials=n_trials, signal_kwargs={'params':params},
                                   bckg_kwargs =dict(atm_keys=['numu_conv'], astro_bckg=astro_bckg), 
                                                           integrate=True, return_data=True)

            inj_trials = ana_single_pl.do_trials(n_trials=n_trials, data=data,
                                   bckg_kwargs =dict(atm_keys=['numu_conv'], astro_bckg=astro_bckg), integrate=True)

        else:
            data, inj_two_comp_i = ana_two_comp_pl.do_trials(n_trials=n_trials,  signal_kwargs={'params':params},
                                   bckg_kwargs =dict(atm_keys=['numu_conv'], astro_bckg=astro_bckg), 
                                                        integrate=True, return_data=True)

            inj_i = ana_single_pl.do_trials(n_trials=n_trials, data=data,
                                   bckg_kwargs =dict(atm_keys=['numu_conv'], astro_bckg=astro_bckg), integrate=True)

            for exp_key in bckg_trials.keys():
                inj_trials[exp_key] = np.append(inj_trials[exp_key], inj_i[exp_key])
                inj_trials_two_comp[exp_key] = np.append(inj_trials_two_comp[exp_key], 
                                                          inj_two_comp_i[exp_key])
        pbar.increment()
    
    RES[(phi0_inji, g_0i)] = (inj_trials_two_comp, inj_trials)
    pbar.finish()
    ppbar.increment()
    
    
ppbar.finish()
```

# Show how well the parameters are recovered

```python
fs=16

median_ts = dict()
for exp_key, trials in bckg_trials.items():
    median_ts[exp_key] = list()

for (phi0_inji, g_0i), (trials0, trials1) in RES.items():
        
    fig, axes = plt.subplots(ncols=5,nrows=2,  figsize=(9,5.), sharex='col')
    fig.suptitle(r'$\gamma$={0:.1f}'.format( g_0i))
    print(15*'#'+' {0:.2f} '.format(g_0i) + 15*'#')
    
    for k, (key,triali) in enumerate(trials1.items()):
        
        
        ################## ts single pl #################
        ax = axes[k, 0]
        bins = np.linspace(0,25, 35)
        ax.hist(triali['ts'], bins=25, histtype='step')
        ax.set_yscale('log')
        if k==1:
            ax.set_xlabel(r'$\Lambda_{\mathrm{single}}$', fontsize=fs)
        ax.set_ylabel(r'pdf', fontsize=fs)

        ################## ts two component pl #################
        ax = axes[k, 1]
        bins = np.linspace(0,25, 35)
        ax.hist(trials0[key]['ts'], bins=25, histtype='step')
        ax.set_yscale('log')

        if k==1:
            ax.set_xlabel(r'$\Lambda_{\mathrm{broken}}$', fontsize=fs)
        ax.set_ylabel(r'pdf', fontsize=fs)

        ################## gamma single pl #################
        ax = axes[k, 2]
        bins = np.linspace(np.min(gammas), np.max(gammas), 30)
        ax.hist(triali['gamma'], bins=bins, histtype='step')
        ax.axvline(g_0i, color='blue', alpha=0.5)
        ax.axvline(gamma_exp, color='orange', alpha=0.5)
        if k==1:
            ax.set_xlabel(r'$\gamma_{\mathrm{single}}$', fontsize=fs)

        
        ################## gammas broken pl #################
        ax = axes[k, 3]
        bins = np.linspace(np.min(gammas), np.max(gammas), 30)
        ax.hist(trials0[key]['gamma0'], bins=bins, histtype='step', color='blue')
        ax.axvline(g_0i, color='blue', alpha=0.5)

        ax.hist(trials0[key]['gamma1'], bins=bins, histtype='step', color='orange')
        ax.axvline(gamma_exp, color='orange', alpha=0.5)
        if k==1:
            ax.set_xlabel(r'$\gamma_{\mathrm{broken}}$', fontsize=fs)

        
        ################## ts difference between both #################
        ax = axes[k, 4]
        bins = np.linspace(0,25, 35)
        ax.hist(trials0[key]['ts'] - triali['ts'] , bins=25, histtype='step')

        med = np.median(trials0[key]['ts'] - triali['ts'])
        median_ts[key].append(med)
        ax.axvline(med)
        
        ax.set_yscale('log')
        if k==1:
            ax.set_xlabel(r'$\Delta \Lambda$', fontsize=fs)
        ax.set_ylabel(r'pdf', fontsize=fs)

        print(np.median(trials0[key]['ts'] - triali['ts']))

    fig.tight_layout()
```

```python
n_trials = int(50)

N = 10

ppbar = ProgressBar(len(g_inj0s),parent=None).start()
# reset random number generator to guarantee same trials
ana_single_pl.reset_random_number_generator(seed=0)
ana_cutoff_pl.reset_random_number_generator(seed=0)

astro_bckg=False
phi0_exp, gamma_exp = 1.44e-18, 2.28


RES = dict()
for i, coffi in enumerate(Cutoff_vals):
    

    params= [phi0_exp, gamma_exp, coffi]

    pbar = ProgressBar(N,parent=ppbar).start()
    for k in range(N):

        if k ==0:
            data, inj_trials_co = ana_cutoff_pl.do_trials(n_trials=n_trials, signal_kwargs={'params':params},
                                   bckg_kwargs =dict(atm_keys=['numu_conv'], astro_bckg=astro_bckg), 
                                                           integrate=True, return_data=True)

            inj_trials = ana_single_pl.do_trials(n_trials=n_trials, data=data,
                                   bckg_kwargs =dict(atm_keys=['numu_conv'], astro_bckg=astro_bckg), integrate=True)

        else:
            data, inj_co_i = ana_cutoff_pl.do_trials(n_trials=n_trials,  signal_kwargs={'params':params},
                                   bckg_kwargs =dict(atm_keys=['numu_conv'], astro_bckg=astro_bckg), 
                                                        integrate=True, return_data=True)

            inj_i = ana_single_pl.do_trials(n_trials=n_trials, data=data,
                                   bckg_kwargs =dict(atm_keys=['numu_conv'], astro_bckg=astro_bckg), integrate=True)

            for exp_key in bckg_trials.keys():
                inj_trials[exp_key] = np.append(inj_trials[exp_key], inj_i[exp_key])
                inj_trials_co[exp_key] = np.append(inj_trials_co[exp_key], 
                                                          inj_co_i[exp_key])
        pbar.increment()
    
    RES[coffi] = (inj_trials_co, inj_trials)
    pbar.finish()
    ppbar.increment()
    
    
ppbar.finish()
```
